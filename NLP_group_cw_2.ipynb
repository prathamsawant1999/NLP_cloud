#### As we are passing the CV data to the NER model, we can't just send in the PDF form of CV to the model, its not going to work. So this helper function below takes in the pdf file, converts the pdf file into a plain text file which is much easier to parse and send the data to the model. 

#### Using fitz we can convert the pdf to text file. The data returned back from this function is json formate with data being key and value is the long string which was parsed.

```python
import fitz
import json as JSON

def pdf_txt(filename, tag):
  docs = fitz.open(filename)
  my_cv = ""
  for doc in docs:
    my_cv = my_cv + str(doc.get_text())
  text_json = {"data": my_cv, "tag": tag}
  pdf_json = JSON.dumps(text_json)
  
  json_data = JSON.loads(pdf_json)
  return json_data
```

#### Another helper function below let us try to manupulate the results returned from the NER API. 
#### The NER model perfomed well recognizing the entity based on the data passed. As some of the CV might classify the education's profile as College Details and some CV education profile might be considered as Degree. So resolve this issue this helper function wil change all the Degree tags to College Name tag as it would be easy for further text process.


```python
#Degree and College name converted to DEGREE
def get_tagged_data(data_str, tag):
  tagged = JSON.loads(data_str)
  tag = tag.upper()

  if tag == 'COLLEGE NAME':
    if "DEGREE" in tagged.keys():
      tagged["COLLEGE NAME"] = tagged.pop("DEGREE")
    return tagged.get(tag)

  else:
    return tagged.get(tag)
```
